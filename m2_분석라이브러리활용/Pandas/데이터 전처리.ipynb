```python
import numpy as np
import pandas as pd
import datetime
id_=np.arange(1,10001)

cusno=np.random.randint(1,1001,size=10000)
gender = np.random.randint(2,size=1000)
age = np.random.randint(81,size=1000)
region=np.random.randint(1,11,size=1000)
c_gen=[]
c_age=[]
c_reg=[]
for i in range(10000):
    c_gen.append(gender[cusno[i]-1])
    c_age.append(age[cusno[i]-1])
    c_reg.append(region[cusno[i]-1])
price = np.random.randint(100,500,size=10000)*100
product = np.random.randint(10,40, size=10000)
quantity = np.random.randint(1,101,size=10000)
time = np.random.randint(1,25,size=10000)
date=pd.date_range(start = '2020-01-01', end = '2021-12-31',freq='D')
datec=np.random.choice(date,10000)

df=pd.DataFrame({'id':id_,'cusno':cusno,'gender':c_gen,'age':c_age,'region':c_reg,'product':product,
                 'price':price,'quantity':quantity,'time':time,
                'date':datec})
df['day']=df.date.dt.day_name()
df['amount']=df.price * df.quantity

ndf=df.reindex(columns=['id','cusno','gender','age','region','product','price','quantity','amount',
                   'date','day','time'])

bin_name=['10','20','30']
hist0,bins0=np.histogram(ndf['price'],bins=3)
ndf['product'] = pd.cut(ndf['price'],
                          bins=bins0,
                          labels=bin_name,
                          include_lowest=True)
ndf.head()
```
