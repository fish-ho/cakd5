```python
# age 칼럼의 범주화
def get_catrgory(age):
    cat=''
    if age <= 5: cat='Baby'
    elif age <= 12: cat='Child'
    elif age <=19: cat='Teenager'
    elif age <=35: cat='Young_adult'
    elif age <=60: cat='Adult'
    else: cat='Elderly'
    return cat

tdf['age_cat'] = tdf.age.apply(lambda x: get_catrgory(x))
tdf[['age','age_cat']].head()
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
print(boston.keys())
```

```python
df_org = pd.DataFrame(boston.data, columns=[boston.feature_names])
df_org['target']=boston.target
df_org.head()
```

```python
df = df_org[['INDUS','RM','PTRATIO','LSTAT','target']]
df.head()
```

```python
# 스케일링
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['RM'] = scaler.fit_transform(df.RM)
df.head()
```

```python
li=['INDUS','RM','PTRATIO','LSTAT','target']
scaler=MinMaxScaler()
for i in li:
    df[i]=scaler.fit_transform(df[i])
df.head()
```
