```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./dataset/auto-mpg.csv')
df.head()
```

```python
# [과제] 'mpg'를 'kpl'로 환산하여 새로운 열을 생성하고 처음 3개 행을 소수점 아래 둘째 자리에서 반올림하여 출력하세요.
df['kpl']=df.mpg*0.425144
df.loc[:2,'kpl']=df.loc[:2,'kpl'].round(2)
df.head()
```

```python
# [과제] horsepower 열을 체크해서 전처리가 필요한 사항을 기술하세요(자료형 타입이 문자로 되어있는 이유 및 float로 처리)
# hint : np.nan으로 대체
# 자료형이 문자형인 이유 -> ?로 채워진 값이 존재 

df=df.replace('?',np.nan)
df['horsepower']=df['horsepower'].astype('float64')
df_h = df.copy()
df.head()
```

```python
# [과제] horsepower 열의 자료형 타입 문제를 해결하고 NAN 값의 개수를 출력하세요
df.horsepower.isnull().sum()
```

```python
# [과제] auto_df를 바이너리 파일로 auto_df.pkl로 저장 후 다시 불러오세요.
df.to_pickle('./dataset/auto_df.pkl')
df1 = pd.read_pickle('./dataset/auto_df.pkl')
df1.head()
```

```python
# [과제] horsepower 열의 최대값의 절대값으로 horsepower 열의 모든 데이터를 나눠서 저장하세요.
abs_h=abs(df1.horsepower.max())
abs_h
df1.horsepower=df1.horsepower/abs_h
df1.head()
```

```python
# [과제] horsepower 각 열 데이터에서 해당 열의 최소값을 뺀 값을 분자, 해당 열의 최대값-최소값을 분모로 계산하여 저장하세요.
min_h=df1.horsepower.min()
max_h=df1.horsepower.max()
num=df1.horsepower - min_h
denom=max_h - min_h
df1.horsepower=num / denom
df1.head()
```

```python
# [과제] origin열의 고유값을 출력하세요.
df2=df1.copy()
df2.origin.unique()
```

```python
# [과제] 정수형 데이터를 문자형 데이터로 변환한 후 고유값을 출력하세요(1:'USA',2:'EU',3:'JAPAN')
df2.origin=df.origin.replace({1:'USA',2:'EU',3:'JAPAN'})
df2.origin.unique()
```

```python
# [과제] origin 열의 자료형을 확인하고 범주형(category)으로 변환하여 출력하세요.
# object
df2['origin']=df2.origin.astype('category')
df2.info()
```

```python
# [과제] origin열을 범주형에서 문자열로 변환한 후 자료형을 출력하세요.
df2['origin']=df2.origin.astype('str')
print(df2['origin'].dtype) 
```

```python
# [과제] horsepower를 3개의 bin으로 나누고 bin_names = ['저출력', '보통출력', '고출력']으로 출력하세요.
# np.histogram() 함수로 3개의 bin으로 나눔
hist,bins=np.histogram(df_h['horsepower'],bins=3) #61.333
print(hist,'\n',bins) # hist는 구간별 개수 , bins는 구간 양끝이 최소값 , 최대값
bin_names = ['저출력', '보통출력', '고출력']
df_h['power'] = pd.cut(df_h['horsepower'],bins = bins,labels = bin_names,include_lowest = True)
print(df_h['power'].value_counts())
df_h
```

```python
# [과제] horsepower 열의 최대값의 절대값으로 모든 데이터를 나누는 정규화를 수행한 후 통계요약표를 출력하세요.
df_h['horsepower']=df_h['horsepower'] / abs(df_h['horsepower'].max())
df_h.describe()
```
