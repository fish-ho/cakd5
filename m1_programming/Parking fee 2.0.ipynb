```python
# [도전 과제] 예외처리 3건 이상 사용 자유 주제

import datetime as dt
import time
dic={}
flag=True
plag=True
base_fee = 3000
fee_per_10min = 500
final_fee=0
able_area=3
def enter():
    try:
        car_num_come= input('차량 번호 입력: ')
        print(len(car_num_come))
        if len(car_num_come) == 7:
            global able_area
            global enter_time
            global e_t
            enter_time=dt.datetime.now()
            e_t=enter_time.strftime('%Y%m%d%H%M%S')
            dic[car_num_come]=e_t
            able_area -= 1
            print(f'주차 현황 - {dic}')
            print(f'주차 가능 공간 : {able_area}')
        else:
            raise
        
    except:
        print('차량번호를 "00가0000" 형식으로 입력해 주세요.')

def work():
    global able_area
    while 1: #정산
        car_num_out= input('차량 번호 입력: ')
        if car_num_out in dic:
            out_time= input('출차 시간 입력: ')
            o_t_for_cal=dt.datetime.strptime(out_time,'%Y%m%d%H%M%S')
            e_t_for_cal=dt.datetime.strptime(e_t,'%Y%m%d%H%M%S')
            dif=o_t_for_cal - e_t_for_cal
            time_total_sec=round(dif.total_seconds())
            time_h=time_total_sec//3600
            time_m=time_total_sec%3600//60
            time_10m=time_m//10

            if time_h > 0:
                final_fee=base_fee + ((time_h-1)*fee_per_10min)*6 + ((time_10m)*fee_per_10min) 
                print(f'{car_num_out} 차량의 주차시간은 {time_h}시간 {time_m}분 \n요금은 {final_fee}원 입니다.')
                while 1:
                    use_coupon=input('주차권(4시간)을 사용하시겠습니까? -- "Y" or "N"')
                    try:
                        if use_coupon =='Y':                       
                            if time_h >= 4:
                                final_fee=((time_h-4)*fee_per_10min)*6 + ((time_10m)*fee_per_10min)
                                print(f'최종 요금은 {final_fee}원 입니다.')
                                pay_fee=input('결제 하시겠습니까? -- "Y" or "N"')
                                if pay_fee =='Y':
                                    print('정산이 완료되었습니다.')
                                    del dic[car_num_out]
                                    able_area += 1
                                    flag=False
                                    plag=False
                                    break

                                elif pay_fee =='N':
                                    print('초기화면으로 돌아갑니다.')
                                    flag=False
                                    plag=False
                                    break

                            else:
                                print('정산이 완료되었습니다.')
                                del dic[car_num_out]
                                able_area += 1
                                flag=False
                                plag=False
                                break

                        elif use_coupon =='N':
                            print(f'요금은 {final_fee}원 입니다. ')
                            pay_fee=input('결제 하시겠습니까? -- "Y" or "N"')
                            if pay_fee =='Y':
                                print('정산이 완료되었습니다.')
                                del dic[car_num_out]
                                able_area += 1
                                flag=False
                                plag=False
                                break

                            elif pay_fee =='N':
                                print('초기화면으로 돌아갑니다.')
                                flag=False
                                plag=False
                                break
                        else:
                            raise
                        
                        if flag == False:
                            break
                        if plag == False:
                            break
                    except Exception as e:
                        #print(e)
                        print('"Y", "N"만 입력하세요')
                    

            else:
                final_fee=base_fee
                print(f'{car_num_out} 차량의 주차시간은 {time_h}시간 {time_m}분 \n 요금은 {final_fee}원 입니다.')
                pay_fee=input('결제 하시겠습니까? -- "Y" or "N"')
                if pay_fee =='Y':
                    print('정산이 완료되었습니다.')
                    del dic[car_num_out]
                    able_area += 1
                    flag=False
                    plag=False
                    break

                elif pay_fee =='N':
                    print('초기화면으로 돌아갑니다.')
                    break

        else:
            print('없는 차량입니다. 차량 번호를 확인하여 주세요.')
            break
            
        if plag == False:
            break

while 1:
    if able_area > 0:
        try:
            work_choice=int(input('1. 입장 정보 입력 2. 정산 3. 종료'))
            if work_choice == 1:
                enter()

            elif work_choice == 2:
                work()

            elif work_choice == 3: # 종료
                break
            else:
                raise
        except:
            print('1,2,3 만 입력하세요')
            
            
    else:
        try:
            work_choice=int(input('2. 정산 3. 종료'))
            if work_choice == 2:
                work()

            elif work_choice == 3: # 종료
                break
            else:
                raise
        except:
            print('2,3 만 입력하세요')
            
        


        
```
